```python
sns.pairplot(insurance,hue = 'smoker')
```

```python
da = insurance[['charges','age', 'bmi', 'children'] and insurance['smoker'] == 'yes']
sns.pairplot(da, kind = 'reg',hue= 'smoker')

##________correlation heat map
print(da.corr())
  
# plotting correlation heatmap
sns.heatmap(da.corr(), cmap="YlGnBu", annot=True)
```
